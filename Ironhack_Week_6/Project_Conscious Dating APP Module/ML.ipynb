```python
import pandas as pd

df = pd.read_csv('/Users/selinwork/Documents/Ironhack/Ironhack_Week_6/Project_Conscious Dating APP Module/Final_Combined_Dating_Dataset_Updated.csv')
```

```python
print(df['ConsciousModule'].value_counts())
```

```python
from sklearn.preprocessing import LabelEncoder

categorical_columns = df.select_dtypes(include=['object']).columns
for col in categorical_columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
```

```python
df.head()
```

```python
categorical_columns = df.select_dtypes(include=['object', 'category', 'int']).columns

# Check unique values for each column
for col in categorical_columns:
    print(f"Column: {col}, Unique Values: {df[col].nunique()}")
```

```python
valid_categorical_columns = [col for col in categorical_columns if df[col].nunique() > 1]

df = df.dropna(subset=valid_categorical_columns)
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(columns=['ConsciousModule'])
y = df['ConsciousModule']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
y.value_counts()
```

```python
from sklearn.ensemble import RandomForestClassifier

RF_model = RandomForestClassifier()
RF_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report

RF_predictions = RF_model.predict(X_test)

print(classification_report(y_test, RF_predictions))
```

```python
print("Test data accuracy: ", RF_model.score(X_test, y_test))
print("Train data accuracy: ", RF_model.score(X_train, y_train))
```

```python
from xgboost import XGBClassifier

# Train XGBoost
xgb_model = XGBClassifier(random_state=42)
xgb_model.fit(X_train, y_train)

# Predict and evaluate
xgb_pred = xgb_model.predict(X_test)

print("XGBoost Classification Report:\n", classification_report(y_test, xgb_pred))
```

```python
xgb_model = XGBClassifier(
    learning_rate=0.1, 
    max_depth=5, 
    n_estimators=100, 
    random_state=42
)
xgb_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

xgb_pred = xgb_model.predict(X_test)
print("XGBoost Accuracy:", accuracy_score(y_test, xgb_pred))
print("XGBoost Classification Report:\n", classification_report(y_test, xgb_pred))
```

```python
from sklearn.naive_bayes import GaussianNB

# Train Naive Bayes
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)

# Predict and evaluate
nb_pred = nb_model.predict(X_test)
print("Naive Bayes Accuracy:", accuracy_score(y_test, nb_pred))
print("Naive Bayes Classification Report:\n", classification_report(y_test, nb_pred))
```

```python
from sklearn.linear_model import LogisticRegression

logistic = LogisticRegression()
logistic.fit(X_train, y_train)

logistic_pred = logistic.predict(X_test)
print("Logistic Regression Accuracy:", accuracy_score(y_test, logistic_pred))
print("Logistic Regression Classification Report:\n", classification_report(y_test, logistic_pred))
```

```python
# Check if the columns exist in the dataframe before dropping them
columns_to_drop = ["Time", "IntentToUse", "ReviewTime", "UsedApps"]
existing_columns_to_drop = [col for col in columns_to_drop if col in df.columns]

df_new_features = df.drop(columns=existing_columns_to_drop)
df_new_features.head()
```

```python
from sklearn.datasets import make_classification

# Generate synthetic data
X_synthetic, y_synthetic = make_classification(
    n_samples=500,  # Number of new samples
    n_features=df.shape[1] - 1,  # Number of features excluding the target
    n_classes=len(df['ConsciousModule'].unique()),  # Number of target classes
    n_informative=10,  # Number of informative features
    random_state=42
)

# Convert to DataFrame
synthetic_data = pd.DataFrame(X_synthetic, columns=df.columns[:-1])
synthetic_data['ConsciousModule'] = y_synthetic

# Combine with original data
augmented_data = pd.concat([df, synthetic_data])

print("Original dataset size:", df.shape)
print("Augmented dataset size:", augmented_data.shape)
```

```python
# Ensure the feature names match those used during training
X = augmented_data.drop(columns=['ConsciousModule'])
y = augmented_data['ConsciousModule']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train Naive Bayes
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)

# Predict and evaluate
nb_pred = nb_model.predict(X_test)
print("Naive Bayes_augmented Accuracy:", accuracy_score(y_test, nb_pred))
print("Naive Bayes Classification Report:\n", classification_report(y_test, nb_pred))
```
